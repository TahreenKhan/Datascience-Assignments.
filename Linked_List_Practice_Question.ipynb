```python
# 1.Definge a doubly linked list[will be done in the class]

# Class for Node in Doubly Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

# Class for Doubly Linked List
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Function to insert a node at the end
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp

    # Function to print the list
    def display(self):
        if not self.head:
            print("List is empty")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()

dll = DoublyLinkedList()
dll.insert_at_end(1)
dll.insert_at_end(2)
dll.insert_at_end(3)
dll.display()  # Output: 1 2 3
```

```python
# 2. Write a funtion to reverse a linked list in-place

# Class for Node in Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Class for Singly Linked List
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # Function to reverse the linked list
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Function to insert a node at the end
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Function to print the list
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()

        
sll = SinglyLinkedList()
sll.insert_at_end(1)
sll.insert_at_end(2)
sll.insert_at_end(3)
sll.display()  
sll.reverse()
sll.display()  
```

```python
# 3. detect cycle in linked list.

# Function to detect cycle in linked list
def detect_cycle(head):
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

# Assuming 'sll' from previous code
# Creating a cycle for testing
sll.head.next.next.next = sll.head.next  # Cycle

# Detecting cycle
print(detect_cycle(sll.head))  


# This function detects a cycle in a singly linked list using Floyd’s Cycle-Finding Algorithm (Tortoise and Hare algorithm):
```

```python
# 4. Merge two sorted linked list into one
# 1->3->5->6->null and 2->4->6->8->null should be merged to make
# 1->2->3->4->5->6->7->8

# Class for Node in Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Class for Singly Linked List
class LinkedList:
    def __init__(self):
        self.head = None

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Function to display the list
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end="->" if temp.next else "->null")
            temp = temp.next
        print()

# Function to merge two sorted linked lists
def merge_sorted_lists(list1, list2):
    # Create a dummy node to start the merged list
    dummy = Node(0)
    tail = dummy

    # Pointers for both lists
    l1 = list1.head
    l2 = list2.head

    # Traverse through both lists
    while l1 and l2:
        if l1.data < l2.data:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next

    # Attach the remaining nodes
    if l1:
        tail.next = l1
    if l2:
        tail.next = l2

    # Return the merged list, skipping the dummy node
    merged_list = LinkedList()
    merged_list.head = dummy.next
    return merged_list


# Create two sorted linked lists
list1 = LinkedList()
list1.insert_at_end(1)
list1.insert_at_end(3)
list1.insert_at_end(5)
list1.insert_at_end(7)

list2 = LinkedList()
list2.insert_at_end(2)
list2.insert_at_end(4)
list2.insert_at_end(6)
list2.insert_at_end(8)

# Display both lists
print("List 1:")
list1.display()  
print("List 2:")
list2.display()  

# Merge both lists
merged_list = merge_sorted_lists(list1, list2)

# Display the merged list
print("Merged List:")
merged_list.display()  
```

```python
# 5. Write a function to remove nth node from the end in a linked list
# 1->2->3->4->5->6, removing 2nd node from end will return 1->2->3->4->6

# Class for Node in Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Class for Singly Linked List
class LinkedList:
    def __init__(self):
        self.head = None

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Function to display the list
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end="->" if temp.next else "->null")
            temp = temp.next
        print()

    # Function to remove the nth node from the end
    def remove_nth_from_end(self, n):
        dummy = Node(0)
        dummy.next = self.head
        first = dummy
        second = dummy

        # Move first pointer n+1 steps ahead
        for _ in range(n + 1):
            if not first:
                return self.head
            first = first.next

        # Move both pointers until first reaches the end
        while first:
            first = first.next
            second = second.next

        # Remove the nth node from the end
        second.next = second.next.next

        # Update the head in case the head node was removed
        self.head = dummy.next

        
# Create the linked list
ll = LinkedList()
ll.insert_at_end(1)
ll.insert_at_end(2)
ll.insert_at_end(3)
ll.insert_at_end(4)
ll.insert_at_end(5)
ll.insert_at_end(6)

# Display the original list
print("Original List:")
ll.display()  # Output: 1->2->3->4->5->6->null

# Remove 2nd node from the end
ll.remove_nth_from_end(2)

# Display the list after removing the 2nd node from the end
print("List after removing 2nd node from end:")
ll.display()  

# Explanation:
# The function remove_nth_from_end uses a dummy node to simplify edge cases, such as when the head node needs to be removed.
# Two pointers, first and second, are used. The first pointer is moved n+1 steps ahead, and both pointers are moved until first reaches the end.
# At this point, the second pointer is just before the node to be removed, allowing us to update the next pointer to bypass the target node.
```

```python
# 6. Remove duplicates from a sorted linked list
# 1->2->3->3->4->4->4->5  should be changed to 1->2->3->4->5

# Class for Node in Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Class for Singly Linked List
class LinkedList:
    def __init__(self):
        self.head = None

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Function to display the list
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end="->" if temp.next else "->null")
            temp = temp.next
        print()

    # Function to remove duplicates from a sorted linked list
    def remove_duplicates(self):
        current = self.head
        while current and current.next:
            if current.data == current.next.data:
                current.next = current.next.next  # Skip the duplicate node
            else:
                current = current.next  # Move to the next distinct node

                
# Create the linked list
ll = LinkedList()
ll.insert_at_end(1)
ll.insert_at_end(2)
ll.insert_at_end(3)
ll.insert_at_end(3)
ll.insert_at_end(4)
ll.insert_at_end(4)
ll.insert_at_end(4)
ll.insert_at_end(5)

# Display the original list
print("Original List:")
ll.display()  # Output: 1->2->3->3->4->4->4->5->null

# Remove duplicates
ll.remove_duplicates()

# Display the list after removing duplicates
print("List after removing duplicates:")
ll.display()  

# Explanation:
# Initial Setup: The remove_duplicates function works only for sorted linked lists, as duplicates will be adjacent.
# Traversal: We traverse the list and check each node with its next node. If they have the same value, the next pointer of the current node is updated to skip the duplicate node.
# No Additional Space: The operation is performed in-place without using extra space.
```

```python
# 7 Find the intersection of two linked list
# 1->2->3->4->8->6->9  5->1->6->7  , intersection 1->6

# Class for Node in Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Class for Singly Linked List
class LinkedList:
    def __init__(self):
        self.head = None

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Function to display the list
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end="->" if temp.next else "->null")
            temp = temp.next
        print()

# Function to find the intersection of two linked lists
def get_intersection_node(head1, head2):
    # Helper function to find the length of a linked list
    def get_length(head):
        length = 0
        current = head
        while current:
            length += 1
            current = current.next
        return length

    # Get lengths of both lists
    len1 = get_length(head1)
    len2 = get_length(head2)

    # Adjust the starting point of the longer list
    if len1 > len2:
        for _ in range(len1 - len2):
            head1 = head1.next
    elif len2 > len1:
        for _ in range(len2 - len1):
            head2 = head2.next

    # Traverse both lists together to find the intersection point
    while head1 and head2:
        if head1 == head2:  # Intersection found
            return head1.data
        head1 = head1.next
        head2 = head2.next

    return None  # No intersection found

# Create two linked lists with intersection
ll1 = LinkedList()
ll2 = LinkedList()

# List 1: 1->2->3->4->8->6->9
ll1.insert_at_end(1)
ll1.insert_at_end(2)
ll1.insert_at_end(3)
ll1.insert_at_end(4)
common_node = Node(8)
ll1.head.next.next.next.next = common_node
common_node.next = Node(6)
common_node.next.next = Node(9)

# List 2: 5->1->6->7 (intersection happens at node with value 6)
ll2.insert_at_end(5)
ll2.head.next = Node(1)
ll2.head.next.next = common_node  # Intersecting node

# Display both lists
print("List 1:")
ll1.display() 
print("List 2:")
ll2.display() 

# Find intersection
intersection_value = get_intersection_node(ll1.head, ll2.head)
if intersection_value:
    print(f"Intersection at node with value: {intersection_value}")
else:
    print("No intersection found")


# xplanation:
# Length Calculation: The function first calculates the length of both linked lists.
# Aligning the Lists: If one list is longer, it advances the pointer in that list to align both lists in terms of distance from the end.
# Traversal to Find Intersection: Both lists are traversed together until an intersection is found or until the end is reached. The intersection node is the one where both pointers reference the same node.
# This method ensures that the algorithm works in O(m + n) time, where m and n are the lengths of the two linked lists, and O(1) extra space.
```

```python
# 8. Rotate a linked list by k positions to right
 # 1->2->3->4->8->6->9 , after rotating for 2 times Cecomes , 3->4->8->6->9->1->2
    
    # Class for Node in Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Class for Singly Linked List
class LinkedList:
    def __init__(self):
        self.head = None

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Function to display the list
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end="->" if temp.next else "->null")
            temp = temp.next
        print()

    # Function to rotate the list by k positions
    def rotate_right(self, k):
        if not self.head or k == 0:
            return

        # Step 1: Find the length of the list and the last node
        length = 1
        tail = self.head
        while tail.next:
            tail = tail.next
            length += 1

        # Step 2: Adjust k (handle cases where k > length)
        k = k % length
        if k == 0:
            return  # No rotation needed if k is 0 or a multiple of the list length

        # Step 3: Make the list circular
        tail.next = self.head

        # Step 4: Traverse to the (length - k)-th node
        new_tail = self.head
        for _ in range(length - k - 1):
            new_tail = new_tail.next

        # Step 5: Set the new head and break the loop
        self.head = new_tail.next
        new_tail.next = None

        
# Create the linked list
ll = LinkedList()
ll.insert_at_end(1)
ll.insert_at_end(2)
ll.insert_at_end(3)
ll.insert_at_end(4)
ll.insert_at_end(8)
ll.insert_at_end(6)
ll.insert_at_end(9)

# Display the original list
print("Original List:")
ll.display()  

# Rotate the list by 2 positions
ll.rotate_right(2)

# Display the rotated list
print("List after rotating by 2 positions:")
ll.display()  


# Explanation:
# Length Calculation: We traverse the list to find the length and the last node.
# Circular Linking: The last node is linked back to the head, creating a circular linked list.
# Find New Head: We compute the new head by moving (length - k) steps from the current head.
# Break the Loop: The node before the new head becomes the new tail, and its next is set to None, breaking the circular link.
# This approach runs in O(n) time, where n is the number of nodes in the list, and uses O(1) extra space.
```

```python
# 9. Add two numbers represented by linked lists:
# Given two non-empty linked lists representing two non-negative integers, where the digits are stored in
# reverse order, add the two numCers and return it as a linked list.

# Class for Node in Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Class for Singly Linked List
class LinkedList:
    def __init__(self):
        self.head = None

    # Function to insert a node at the end of the list
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Function to display the list
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end="->" if temp.next else "->null")
            temp = temp.next
        print()

# Function to add two numbers represented by linked lists
def add_two_numbers(l1, l2):
    dummy = Node(0)  # Dummy node to keep track of the head of the result list
    current = dummy  # Pointer to build the result list
    carry = 0  # To store carry from each sum

    # Traverse both lists
    while l1 or l2:
        x = l1.data if l1 else 0  # Take value from l1 or 0 if l1 is None
        y = l2.data if l2 else 0  # Take value from l2 or 0 if l2 is None
        total = carry + x + y  # Sum current digits and carry
        carry = total // 10  # Update carry
        current.next = Node(total % 10)  # Create new node with the remainder of the sum
        current = current.next  # Move to the next node

        # Move to the next digits in l1 and l2 if possible
        if l1: l1 = l1.next
        if l2: l2 = l2.next

    # If there's a carry left after the final addition, add a new node
    if carry:
        current.next = Node(carry)

    return dummy.next  # Return the next node after dummy as the head of the result list

# Create first number: 243 (3 -> 4 -> 2 in reverse order)
l1 = LinkedList()
l1.insert_at_end(3)
l1.insert_at_end(4)
l1.insert_at_end(2)

# Create second number: 465 (5 -> 6 -> 4 in reverse order)
l2 = LinkedList()
l2.insert_at_end(5)
l2.insert_at_end(6)
l2.insert_at_end(4)

# Display both numbers
print("Number 1:")
l1.display() 
l2.display() 

# Add the two numbers
result = add_two_numbers(l1.head, l2.head)

# Display the result (sum 708, represented as 8 -> 0 -> 7 in reverse order)
print("Sum:")
result_list = LinkedList()
result_list.head = result
result_list.display()


# Explanation:
# Traversal of Lists: We traverse both linked lists simultaneously. For each pair of corresponding digits, we add them along with any carry from the previous sum.
# Carry Handling: We use integer division (// 10) to compute the carry and modulus (% 10) to get the current digit.
# Handling Different Lengths: If one list ends before the other, we continue adding the remaining digits from the longer list.
# Final Carry: If there's still a carry after adding all digits, we append it as the last node in the result.
# This algorithm runs in O(max(m, n)) time, where m and n are the lengths of the two linked lists, and uses O(max(m, n)) space for the result list.
```
